```python
from bs4 import BeautifulSoup
```

```python
def strip_html_tags(text):
    soup=BeautifulSoup(text, "html.parser")
    [s.extract() for s in soup(['iframe','script'])]
    stripped_text=soup.get_text()
    stripped_text=re.sub(r'[\r|\n|\r\n]+','\n',stripped_text)
    return stripped_text
```

```python
clean_content=strip_html_tags(content)
```

```python
def tokenize_text(text):
    sentences=nltk.sent_tokenize(text)
    word_tokens=[nltk.word_tokenize(word) for word in sentences]
    return word_tokens
```

```python
sents=tokenize_text(sample_text)
```

```python
np.array(tokenize_text(sample_text))
```

```python
words=[word for sentence in sents for word in sentence]
```

```python
import unicodedata
```

```python
def remove_accented_chars(text):
    text=unicodedata.normalize('NKFD', text).encode('ascii','ignore').decode('utf-8','ignore')
    return text
```

```python
remove_accented_chars('Some Accented Txt')
```
