```python
import re, collections
```

```python
def tokens(text):
    
    return re.findall('[a-z]+', text.lower())
```

```python
words=tokens(open('big.txt').read())
```

```python
word_counts=collections.Counter(words)
```

```python
word_counts.most_common(10)
```

```python
def edits0(word):
    return {word}
```

```python
def edits1(word):
    alphabet='abcdefghijklmnopqrstuvwxyz'
    def split(word):
        return[(word[:i], word[i:]) for i in range (len(word)+1)]
    pairs=split(word)
    deletes=[a+b[1:] for (a,b) in pairs if b]
    transposes=[a+b[1]+b[0]+b[2:] for (a,b) in pairs if len(b)>1]
    replaces=[a+c+b[1:] for (a,b) in pairs for c in alphabet if b]
    inserts=[a+c+b for (a,b) in pairs for c in alphabet]
    return set(deletes+transposes+replaces+inserts)
```

```python
def edits2(word):
    return{e2 for e1 in edits1(word) for e2 in edits1(e1)}
```

```python
def known(words):
    return{w for w in words if w in word_counts}
```

```python
word='fianlly'
```

```python
edits0(word)
```

```python
known(edits0(word))
```

```python
edits1(word)
```

```python
known(edits1(word))
```

```python
edits2(word)
```

```python
known(edits2(word))
```

```python
candidates=(known(edits0(word)) or known(edits1(word)) or known(edits2(word)) or [word])
```

```python
candidates
```

```python
def correct(word):
    candidates=(known(edits0(word)) or known(edits1(word)) or known(edits2(word)) or [word])
    return max(candidates, key=word_counts.get)
```

```python
correct('fianlly')
```

```python
correct('FIANLLY')
```

```python
def correct_match(match):
    word=match.group()
    def case_of(text):
        return(str.upper if text.isupper() else str.lower if text.islower() else str.title if text.istitle() else str)
    return case_of(word)(correct(word.lower()))
```

```python
def correct_text_generic(text):
    return(re.sub('[a-zA-Z]+', correct_match, text))
```

```python
correct_text_generic('fiannlly')
```

```python
correct_text_generic('pleyful')
```
